# 自定义模块
新建py文件 然后导入即可\
如果不同模块有同名功能，则导入的后面的那个是被导入的\


## __main__ 变量

```python
import my_module1
my_module1.test(1,2)

#import的时候就执行了
#解决办法 main
```

```python
#当用from导入的时候，main不会被运行
from my_module1 import test
```
